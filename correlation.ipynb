This notebook calculates the correlation and its significance between the incremental values.

```python
import pickle
from scipy.stats import pearsonr
import numpy as np
```

```python
#Load complete dataset
with open(r"/home/users/fschreiber/project/data_frames/inc_metrics/ReSc/inc/df_all","rb") as f:
    df_full=pickle.load(f)
```

```python
df_full
```

```python
# Calculate the correlation matrix
rho = df_full.corr()

# Calculate p-values for correlation coefficients
pval = df_full.corr(method=lambda x, y: pearsonr(x, y)[1]) - np.eye(*rho.shape)

#* show significant correlation
p = pval.applymap(lambda x: ''.join(['*' for t in [.05, .01, .001] if x <= t]))

rho.round(2).astype(str)

rho.round(2).astype(str) + p
```
